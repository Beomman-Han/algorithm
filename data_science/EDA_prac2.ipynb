```python
%matplotlib inline

import pandas as pd
import seaborn as sns
```

```python
## EDA example with titanic passenger

titanic = pd.read_csv('data/titanic.csv')
titanic
```

```python
titanic.info()
```

```python
titanic.plot(kind='hist', y='Age', bins=30)
```

```python
titanic.plot(kind='scatter', x='Age', y='Fare')
```

```python
titanic['Survived'].value_counts()
```

```python
titanic['Pclass'].value_counts()
```

```python
titanic.plot(kind='scatter', x='Survived', y='Pclass')
```

```python
sns.kdeplot(titanic['Survived'], titanic['Pclass'])
```

```python
#titanic.plot(kind='box', x='Survived', y='Age')
sns.stripplot(data=titanic, x='Survived', y='Age')
```

```python
sns.violinplot(data=titanic, x='Survived', y='Age')
```

```python
sns.catplot(data=titanic, x='Survived', y='Age', kind='box')
```

```python
sns.stripplot(data=titanic, x='Survived', y='Age', hue='Sex')
```
