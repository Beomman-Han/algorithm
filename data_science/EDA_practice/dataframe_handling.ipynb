```python
df = pd.read_csv('data/laptops.csv')
df
```

```python
brand_nations = {
    'Dell' : 'U.S.',
    'Apple' : 'U.S.',
    'Acer' : 'Taiwan',
    'HP' : 'U.S.',
    'Lenovo' : 'China',
    'Alienware' : 'U.S.',
    'Microsoft' : 'U.S.',
    'Asus' : 'Taiwan'
}
```

```python
df['brand']
```

```python
df['brand'].map(brand_nations)  ## **
```

```python
df['brand_nations'] = df['brand'].map(brand_nations)
```

```python
df
```

```python
## groupby **
```

```python
nation_groups = df.groupby('brand_nations')
type(nation_groups)
```

```python
nation_groups.count()
```

```python
nation_groups.mean()
```

```python
nation_groups.first()
```

```python
nation_groups.last()
```

```python
## merging data (***)
```

```python
price_df = pd.read_csv('data/vegetable_price.csv')
quantity_df = pd.read_csv('data/vegetable_quantity.csv')
```

```python
price_df
```

```python
quantity_df
```

```python
pd.merge(price_df, quantity_df, on='Product')  ## inner join (**)
```

```python
pd.merge(price_df, quantity_df, on='Product', how='left')  ## left outer join (**)
```

```python
pd.merge(price_df, quantity_df, on='Product', how='right')  ## right outer join (**)
```

```python
pd.merge(price_df, quantity_df, on='Product', how='outer')  ## full outer join (**)
```
