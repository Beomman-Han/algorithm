```python
import pandas as pd
```

```python
df = pd.read_csv('data/enrolment_1.csv')
df
```

```python
cond1 = (df['year'] == 1) & (df['course name'] == 'information technology')
cond1
```

```python
cond2 = (df['year'] == 4) & (df['course name'] == 'commerce')
cond2
```

```python
df['status'] = 'allowed'
df.loc[(cond1 | cond2), 'status'] = 'not allowed'
df['status'].describe()
```

```python
course_count = df.loc[df['status'] == 'allowed', 'course name'].value_counts()
```

```python
closed = course_count[course_count < 5]
closed.index
```

```python
for course in closed.index:
    df.loc[df['course name'] == course, 'status'] = 'not allowed'
```
