```python
from typing import Tuple
import numpy as np

import matplotlib.pyplot as plt
```

```python
def prediction(
    theta_0 : float,
    theta_1 : float,
    x : np.array
    ) -> np.array:
    
    """Return prediction value of linear regression
    h(x) = theta_0 + theta_1 * x"""
    
    return theta_0 + theta_1 * x
```

```python
def prediction_difference(
    theta_0 : float,
    theta_1 : float,
    x : np.array,
    y : np.array
    ) -> np.array:
    
    """Return difference btw prediction values and
    target values (error)"""
    
    return prediction(theta_0, theta_1, x) - y
```

```python
def gradient_descent(
    theta_0 : float,
    theta_1 : float,
    x : np.array,
    y : np.array,
    iterations : int,
    alpha : float
    ) -> Tuple[float, float]:
    
    m = len(x)
    costs = []
    
    for i in range(iterations):
        error = prediction_difference(theta_0, theta_1, x, y)
        
        costs.append((error @ error) / (2 * m))
        
        theta_0 -= alpha * error.mean()
        theta_1 -= alpha * ( error * x ).mean()
        
        ## visualizing
        if i % 10 == 0:
            plt.scatter(house_size, house_price)
            plt.plot(house_size, prediction(theta_0, theta_1, x), color='red')
            plt.show()
    
    return theta_0, theta_1, costs
```
