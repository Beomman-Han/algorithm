```python
## one-hot encoding practice with pandas
import pandas as pd
```

```python
titanic_df = pd.read_csv('./data/titanic.csv', index_col=0)
titanic_df
```

```python
## transform 'Sex', 'Embarked' categorical data to numerical data
titanic_sex_embarked = titanic_df[['Sex', 'Embarked']]
titanic_sex_embarked
```

```python
## one-hot encoding
one_hot_encoded_df = pd.get_dummies(titanic_sex_embarked)
one_hot_encoded_df.head()
```

```python
## one-hot encoding for specific columns from whole dataframe
one_hot_encoded_df = pd.get_dummies(data=titanic_df, columns=['Sex', 'Embarked'])
one_hot_encoded_df.head()
```
